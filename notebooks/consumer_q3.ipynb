```python
import json
from kafka import KafkaConsumer, TopicPartition

consumer = KafkaConsumer(
    bootstrap_servers='localhost:9092',
    auto_offset_reset='earliest',
    value_deserializer=lambda v: json.loads(v.decode('utf-8')),
)

# Assign partition manually so we can check end offsets
tp = TopicPartition('green-trips', 0)
consumer.assign([tp])
consumer.seek_to_beginning(tp)

# Get the last offset so we know exactly when to stop
end_offset = consumer.end_offsets([tp])[tp]
print(f'Total messages in topic: {end_offset}')

count = 0
total = 0

for msg in consumer:
    record = msg.value
    total += 1
    if record.get('trip_distance', 0) > 5:
        count += 1

    # Stop exactly when we've read all messages
    if msg.offset + 1 >= end_offset:
        break

print(f'Total trips consumed: {total}')
print(f'Trips with trip_distance > 5.0: {count}')
```
